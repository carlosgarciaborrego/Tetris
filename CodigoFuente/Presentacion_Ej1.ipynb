```python
#Puede que el mapa, se sustituya por una matriz normal
class Mapa:  
    def __init__(self, celdas):
        self.celdas = celdas
    
    def tamaño_hor(self):
        return len(self.celdas[0])
    
    def tamaño_ver(self):
        return len(self.celdas)
    
    def tipo_celda(self, f, c):
        return self.celdas[f][c]
    
    def __str__(self): # No hace falta, pero sirve para poder imprimir el mapa por pantalla
        res = ""
        
        for celda in self.celdas:
            
            res += str(celda) + "\n"
        
        return res
 
```

```python
def rotacion_izq(estado):
    #Filas (X,)
    x_menor = 99
    #Columnas (,Y)
    y_menor = 99
    
    #Almacenamos el valor más pequeño de la x e y
    for i in range(len(estado)):
        for j in range(2):
            if estado[i][0] < x_menor:
                x_menor = estado[i][0]
            if estado[i][1] < y_menor:
                y_menor = estado[i][1]

    #Matriz de ceros 3x3
    nuevaMatriz = np.zeros((3,3),dtype= int)
    #Copiamos el estado inicial por si no se pudiera realizar con éxito la acción de rotar
    estadoAuxiliar = copy.deepcopy(estado)

    #Hacemos que la matriz 3x3 empiece en el (0,0) hasta el (2,2)
    for i in range (len(estadoAuxiliar)):
        for j in range(2):
            if j == 0:
                estadoAuxiliar[i][0] = estadoAuxiliar[i][0] - x_menor
            if j == 1:
                estadoAuxiliar[i][1] = estadoAuxiliar[i][1] - y_menor
    
    #Colocamos en la nueva matriz(de ceros), con un 4 las posiciones donde van las celdas de la figura/pieza(que haya salido) 
    for i in range (len(estadoAuxiliar)):
        for j in range(len(estadoAuxiliar[i])):
            fila = estadoAuxiliar[i][0]
            columna = estadoAuxiliar[i][1]
            nuevaMatriz[fila][columna] = 4
        
    #Rotamos la nuevaMatriz 90º a la izquierda
    rot1 = np.rot90(nuevaMatriz)

    #Preparamos una lista de posiciones donde almacenaremos las posiciones finales tras realizar la rotación
    posiciones_finales = [[0,0],[0,0],[0,0],[0,0]]

    contador = 0      
    for i in range(3):
        for j in range(3):
            if rot1[i][j] == 4:
                var_x = i + x_menor
                var_y = j + y_menor
                posiciones_finales[contador][0] = var_x
                posiciones_finales[contador][1] = var_y
                contador = contador + 1
    
    return posiciones_finales

def rotacion_der(estado):
    #Filas (X,)
    x_menor = 99
    #Columnas (,Y)
    y_menor = 99
    
    #Almacenamos el valor más pequeño de la x e y
    for i in range(len(estado)):
        for j in range(2):
            if estado[i][0] < x_menor:
                x_menor = estado[i][0]
            if estado[i][1] < y_menor:
                y_menor = estado[i][1]

    #Matriz de ceros 3x3
    nuevaMatriz = np.zeros((3,3),dtype= int)
    #Copiamos el estado inicial por si no se pudiera realizar con éxito la acción de rotar
    estadoAuxiliar = copy.deepcopy(estado)

    #Hacemos que la matriz 3x3 empiece en el (0,0) hasta el (2,2)
    for i in range (len(estadoAuxiliar)):
        for j in range(2):
            if j == 0:
                estadoAuxiliar[i][0] = estadoAuxiliar[i][0] - x_menor
            if j == 1:
                estadoAuxiliar[i][1] = estadoAuxiliar[i][1] - y_menor
    
    #Colocamos en la nueva matriz(de ceros), con un 4 las posiciones donde van las celdas de la figura/pieza(que haya salido) 
    for i in range (len(estadoAuxiliar)):
        for j in range(len(estadoAuxiliar[i])):
            fila = estadoAuxiliar[i][0]
            columna = estadoAuxiliar[i][1]
            nuevaMatriz[fila][columna] = 4
        
    #Rotamos 3 veces la nuevaMatriz 90º a la izquierda, lo que corresponde con rotar 1 vez a la derecha
    rot3 = np.rot90(nuevaMatriz,3)

    #Preparamos una lista de posiciones donde almacenaremos las posiciones finales tras realizar la rotación
    posiciones_finales = [[0,0],[0,0],[0,0],[0,0]]

    contador = 0      
    for i in range(3):
        for j in range(3):
            if rot3[i][j] == 4:
                var_x = i + x_menor
                var_y = j + y_menor
                posiciones_finales[contador][0] = var_x
                posiciones_finales[contador][1] = var_y
                contador = contador + 1
    
    return posiciones_finales
```

```python
# Acciones:
prueba = 1
class Abajo(probee.Acción):
    
    def __init__(self):
        nombre = 'Abajo'
        super().__init__(nombre)
        
    def es_aplicable(self,estado):           
        # Hay que comprobar que no estémos en el límite inferior del mapa *and* que la casilla inferior no sea pieza
        res = False
        contador = 0
        for i in range (len(estado)):
            if estado[i][0] < mapa.tamaño_ver() and (estado[i][0]+1  < mapa.tamaño_ver()) and mapa.celdas[estado[i][0] + 1][estado[i][1]] == 0:
                contador +=1
            if contador == 4:
                res = True
           
        return res       
    
    def aplicar(self,estado):
        nuevo_estado = copy.deepcopy(estado)
        #nuevo_estado[0] += 1  # bajamos una fila
        for i in range (len(estado)):
            nuevo_estado[i][0] +=1
            
        return nuevo_estado
    
    def coste_de_aplicar(self,prueba):
        return 1

    

class Izquierda(probee.Acción):
    def __init__(self):
        nombre = 'Izquierda'
        super().__init__(nombre)
        
    def es_aplicable(self,estado):           
        # Hay que comprobar que no estémos en el límite izquierdo del mapa y que no haya piezas a la izquierda
        res = False
        contador = 0
        for i in range (len(estado)):
            if estado[i][1] > -1 and (estado[i][1]-1 > -1) and mapa.celdas[estado[i][0]][estado[i][1]-1] == 0:
                contador += 1
                
            if contador == 4:
                res = True
        
        return res
                
    def aplicar(self,estado):
        nuevo_estado = copy.deepcopy(estado)
        for i in range (len(estado)):
            nuevo_estado[i][1] -=1
            
        return nuevo_estado
    
    def coste_de_aplicar(self,prueba):
        return 1
    
    
class Derecha(probee.Acción):
    def __init__(self):
        nombre = 'Derecha'
        super().__init__(nombre)
        
    def es_aplicable(self,estado):           
        # Hay que comprobar que no estémos en el límite derecho del mapa *and* que la casilla derecha no sea agua
        res = False
        contador = 0
        for i in range (len(estado)):
            if (estado[i][1] < mapa.tamaño_hor()) and (estado[i][1]+1 < mapa.tamaño_hor()) and mapa.celdas[estado[i][0]][estado[i][1]+1] == 0:
                contador += 1
                
            if contador == 4:
                res = True
        
        return res
        
    def aplicar(self,estado):
        nuevo_estado = copy.deepcopy(estado)
        for i in range (len(estado)):
            nuevo_estado[i][1] +=1
           
        return nuevo_estado
    
    def coste_de_aplicar(self,prueba):
        return 1    

    
class Girar_izquierda(probee.Acción):
    def __init__(self):
        nombre = 'Girar_izquierda'
        super().__init__(nombre)
        
    def es_aplicable(self,estado):
        # Hay que comprobar que no estémos en el límite izquierda del mapa *and* que la casilla izquierda no sea 1
        res = False
        contador = 0
        nuevoEstado = rotacion_izq(estado)
        for i in range (len(estado)):
            #Será aplicable si al rotar a la izquierda, las celdas son distintas de 1 y no esta fuera del mapa
            if (nuevoEstado[i][1] > -1) and (nuevoEstado[i][1] < mapa.tamaño_hor()) and  (nuevoEstado[i][0] < mapa.tamaño_ver()) and (nuevoEstado[i][0]>-1) and mapa.celdas[nuevoEstado[i][0]][nuevoEstado[i][1]] == 0:
                contador += 1
            
            if contador == 4:  
                res = True
        
        return res
    
    def aplicar(self,estado):
        nuevo_estado = copy.deepcopy(estado)
        nuevo_estado2 = rotacion_izq(nuevo_estado)
        return nuevo_estado2
    
        
    def coste_de_aplicar(self,prueba):
        return 1    
    

class Girar_derecha(probee.Acción):
    def __init__(self):
        nombre = 'Girar_derecha'
        super().__init__(nombre)
        
    def es_aplicable(self,estado):
        # Hay que comprobar que no estémos en el límite derecha del mapa *and* que la casilla izquierda no sea 1
        res = False
        contador = 0
        nuevoEstado = rotacion_der(estado)
        for i in range (len(estado)):
            #Será aplicable si al rotar a la derecha, las celdas son distintas de 1 y no esta fuera del mapa
            if (nuevoEstado[i][1] > -1) and (nuevoEstado[i][1] < mapa.tamaño_hor()) and  (nuevoEstado[i][0] < mapa.tamaño_ver()) and (nuevoEstado[i][0]>-1) and mapa.celdas[nuevoEstado[i][0]][nuevoEstado[i][1]] == 0:
                contador += 1
                
            if contador == 4:
                res = True
        
        return res
    
    def aplicar(self,estado):
        nuevo_estado = copy.deepcopy(estado)
        nuevo_estado2 = rotacion_der(nuevo_estado)
        return nuevo_estado2
    
        
    def coste_de_aplicar(self,prueba):
        return 1    
      
```

```python
# Método para transformar de Abajo,Izquierda,Derecha,Girar Izquierda,Girar Derecha  a  A,I,D,G_Izq,G_Der respectivamente
def transforma_solucion(solucion):
    res = ""
    for palabra in solucion:
        if(palabra == "Abajo"):
            res += "A "  # Arriba
            continue
        if(palabra == "Izquierda"):
            res += "I "  # Izquierda
            continue
        if(palabra == "Derecha"):
            res += "D "  # Derecha
            continue
        if(palabra == "Girar_izquierda"):
            res += "G_izq "  # Girar_izquierda  
            continue
        if(palabra == "Girar_derecha"):
            res += "G_der "  # Girar_derecha      
    return res            
```

```python
#Métodos pertenecientes al cálculo de la heurística 'h'
#Cuenta el número de celdas que tienen valor 1
def numeroCeldasLlenas(mapa): 
    contador=0
    for i in range (22): 
        for j in range (10):
            if mapa[i][j]==1:
                contador=contador + 1
    return contador    

#Calcula la altura máxima a la que se encuentra los 1
def maximaAlturaAlcanzada(mapa):
    altura=0
    for i in range (22): 
        for j in range (10):
            if mapa[21-i][9-j]==1:
                altura = i
    return altura
      
#Calcula el número de huecos(se considera hueco si tiene a izquierda,derecha y arriba una celda con valor 1) en el mapa_inicial
def buscarHuecos(mapa):
    huecos=0
    for i in range (22): 
        for j in range (10):
            if j==9 and mapa[i][j]==0 and mapa[i-1][j]==1 and mapa[i][j-1]==1:
                huecos = huecos + 1
            if j== 0 and mapa[i][j]==0 and mapa[i-1][j]==1 and mapa[i][j+1]==1:
                huecos = huecos + 1
            if j<9 and j >0 and mapa[i][j]==0 and mapa[i-1][j]==1 and mapa[i][j-1]==1 and mapa[i][j+1]==1: 
                huecos = huecos + 1
    return huecos 

#Suma el valor de la altura de cada 1 que aparece en el mapa
def alturaPonderada(mapa):
    alturaPonderada = 0
    for i in range (22): 
        for j in range (10):
            if mapa[i][j]==1:
                alturaPonderada = alturaPonderada + (22-i)*1
    return alturaPonderada 

#Busca si hay una linea completa con 1's, y penaliza en negativo para que sea el principal objetivo.
def buscarLineas(mapa):
    res = 0
    for i in range(22):
        cont = 0
        if 1 in mapa[i]:
            for j in range(10):
                if mapa[i][j] == 1:
                    cont = cont + 1

            if cont == 10:
                res = -1000
    return res  
```

```python
def h(nodo):
    # Como heurística ejemplo, calculamos la distancia desde la posición del nodo hasta el estado final más cercano
    
    res = None
    estado = nodo.estado
    
    for estado_final in estados_finales:
        h0 = numeroCeldasLlenas(mapa_inicial)
        h1 = maximaAlturaAlcanzada(mapa_inicial)
        h2 = buscarHuecos(mapa_inicial)
        h3 = alturaPonderada(mapa_inicial)
        h4 = buscarLineas(mapa_inicial)
        c0 = 1
        c1 = 1
        c2 = 20
        c3 = 5
        c4 = 1
        aux = c0*h0 + c1*h1 + c2*h2 + c3*h3 + c4*h4
        if res == None: 
            res = aux
        elif aux < res:  
            res = aux

    return res
```
